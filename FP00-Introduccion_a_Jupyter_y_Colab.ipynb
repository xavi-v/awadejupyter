## <font color='blue'>**Utilizando Jupyter/Colab Notebooks**</font>
Cada notebook está compuesto por una serie de celdas de trabajo. Cada celda tiene dos modos de funcionamiento:
- Code Mode
- Markdown (text) mode

Para pasar de un modo a otro debes presionar la tecla ***\<esc\>*** y luego la tecla ***\<Y\>*** (para pasar a modo ***Code***), o la tecla ***\<M\>*** (para pasar a modo ***Markdown***)

La siguiete celda está en como código. Para ejecutarla debes presionar las teclas ***\<Shift + Enter\>***

```python
# Esta es una celda de código; se caracteriza porque a su izquierda tiene un par de corchetes []
print("Hola mundo")
```

### Atajos de teclado en Colab

En **Celda de código**. Una vez que el botón de la barra de herramientas indique CONECTADO, haz clic en la celda para seleccionarla y ejecutar el contenido de las siguientes formas:

* Haz clic en el **icono Reproducir** en el margen izquierdo de la celda;
* Escribe **Cmd / Ctrl + Enter** para ejecutar la celda en su lugar;
* Escribe **Shift + Enter** para ejecutar la celda y mover el foco a la siguiente celda (agregando una si no existe ninguna); o
* Escribe **Alt + Enter** para ejecutar la celda e insertar una nueva celda de código inmediatamente debajo de ella.

Hay opciones adicionales para ejecutar algunas o todas las celdas en el menú **Runtime**.

### Las salidas (outputs) de las celdas son asíncronas
Esto quiere decir que los resultados se ven conforme se van generando en el kernel. Ejecuta el siguiente código (\<Shift+Enter\>)

```python
import time, sys
for i in range(8):
    print(i)
    time.sleep(0.5)
```

### Salidas grandes
Para manejar mejor las salidas grandes, el área de salida se puede contraer. Ejecute la siguiente celda y luego haz clic simple o doble en el área activa a la izquierda de la salida:

```python
for i in range(50):
    print(i)
```

Pero a partir de cierto punto, la salida comienza  a hacer "scrolling" automáticamente. Fíjate en la siguiente celda que en el área de salida aparece un manejador de scrolling a la derecha. Si lo arrastras, verás que la salida es más grande de lo que se ve:

```python
for i in range(500):
    print(2**i - 1)
```
